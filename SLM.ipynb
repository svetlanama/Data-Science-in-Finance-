```python
import matplotlib.pyplot as plt

# Data
metrics = [
    "Customer Wait Time ↓",
    "Autonomous Query Handling",
    "Deflection from Agents",
    "Customer Satisfaction ↑"
]
values = [70, 85, 40, 25]

# Colors
primary_color = "#DA291C"
secondary_color = "#232323"

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(metrics, values, color=primary_color)

# Add value labels
for bar in bars:
    ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
            f"{bar.get_width()}%", va='center', fontsize=12, color=secondary_color)

# Style
ax.set_xlim(0, 100)
ax.set_xlabel("Improvement (%)", fontsize=12)
ax.set_title("Virtual Assistant – Performance Metrics", fontsize=16, color=secondary_color)
ax.invert_yaxis()
ax.set_facecolor("white")
fig.patch.set_facecolor("white")
ax.tick_params(colors=secondary_color)
ax.xaxis.label.set_color(secondary_color)
ax.title.set_fontweight("bold")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Data
metrics = [
    "Customer Wait Time ↓",
    "Autonomous Query Handling",
    "Deflection from Agents",
    "Customer Satisfaction ↑"
]
values = [70, 85, 40, 25]

# Colors (SLM Style)
primary_color = "#DA291C"      # SLM Red
text_color = "#232323"         # Charcoal
bg_color = "#FFFFFF"           # White

# Plot Setup
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(metrics, values, color=primary_color, height=0.5)

# Add labels
for bar in bars:
    ax.text(bar.get_width() + 2, bar.get_y() + bar.get_height()/2,
            f"{int(bar.get_width())}%", va='center', fontsize=13, color=text_color, weight='bold')

# Title and Labels
ax.set_title("Virtual Assistant – Performance Metrics", fontsize=18, weight='bold', color=text_color, pad=20)
ax.set_xlabel("Improvement (%)", fontsize=12, color=text_color)
ax.set_xlim(0, 100)
ax.invert_yaxis()  # Highest metric at top

# Style
ax.set_facecolor(bg_color)
fig.patch.set_facecolor(bg_color)
ax.tick_params(colors=text_color, labelsize=11)
ax.xaxis.label.set_color(text_color)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#dddddd')
ax.spines['bottom'].set_color('#dddddd')
ax.grid(axis='x', linestyle='--', linewidth=0.5, color='#dcdcdc')

# Padding
plt.tight_layout()
plt.show()
```
